```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import requests
import pandas as pd
import numpy as np
```

```python
# getting different sectors
url  = "https://ticker.finology.in/sector"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
}
 
response = requests.get(url, headers=headers)
 
if response.status_code == 200:
    print(response.text)  # Print the content of the response
else:
    print(f'Request failed with status code: {response.status_code}')
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
potential_data  = soup.find("div", {"id": "sectors"}).find_all("div",{"class": "col-12 col-md-3"})
all_links = []
for i in potential_data:
    inner_data = i.find("div", {"class": "card cardsector cardscreen cardsmall text-center"}).find("div",{"class": "mt-4 text-center"})
    href_link  = inner_data.find("a").get("href")
    all_links.append(href_link)
```

```python
# final function
def company_extractor():
    i1 = 300  # also change here 
    company_link  = []
    while(i1 < 500):
        url = "https://ticker.finology.in/Directory/Stocks/{}".format(i1)
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
        response = requests.get(url , headers= headers)
        if(response.status_code == 200):
            # further processing
            soup = BeautifulSoup(response.text, 'html.parser')
            potential_data  = soup.find("form",{"id": "frmTickerMaster"}).find("div",{"class": "innerpagecontent my-5"}).find_all("div", {"class": "col-12 col-md-4"})
            for i in potential_data:
                alfa = i.find("a", {"class": "btn btn-sm btn-primary ml-0 col-4"}).get("href")
                company_link.append(alfa)
                print(alfa)
        
        else:
            # failsafe system generation
            print(f'Request failed with status code: {response.status_code}')
            break  # break condition
        i1 += 1
    return company_link
```

```python
company_list = company_extractor()
```
